3) Os dados a seguir correspondem aos pesos, em kg, dos sacos de um determinado produto. Faça um gráfico BoxPlot e determine:

a) se existe outliers.

b) Q1 e Q3

c) O intervalo interquartil

d) A mediana

1,01 0,85 1,11 1,12 1,14 1,14 1,15 1,15 1,10 1,18 1,12 1,15 0,79 1,01 1,03 1,05 1,08 0,62 1,08 1,09 1,01 1,20 1,21 1,05 1,08 1,09 1,10 1,15 1,12 1,15

```python
import matplotlib.pyplot as plt
import numpy as np
Peso = [0.62, 0.79, 0.85, 1.01, 1.01, 1.01, 1.03, 1.05, 1.05, 1.08,
        1.08, 1.08, 1.08, 1.09, 1.09, 1.10, 1.10, 1.11, 1.12, 1.12,
        1.12, 1.14, 1.14, 1.15, 1.15, 1.15, 1.15, 1.18, 1.20, 1.21]
Q1 = np.percentile(Peso, 25, method = "midpoint")
Q3 = np.percentile(Peso, 75, method = "midpoint")
mediana = np.median(Peso)
I = Q3 - Q1
LI = Q1 - 1.5 * I
LS = Q3 + 1.5 * I

outlier = []
for i in range(len(Peso)):
  if Peso[i] > LS or Peso[i] < LI:
    outlier.append(Peso[i])
print(f"outlier = {outlier}")
print(f"Q1 = {Q1}")
print(f"Q3 = {Q3}")
print(f"I = {I}")
print(f"Máximo = {LS}")
print(f"Mínimo = {LI}")
print(f"mediana = {mediana}")

plt.boxplot(Peso, vert = False)
plt.show()
```

4) Uma turma com 22 alunos realizou uma prova de física e obtiveram os seguintes resultados

7,8 - 5,0 - 3,1 - 4,8 - 4,4 - 5,4 - 8,5 - 5,9 - 5,9 - 3,1 - 5,6 - 3,7 - 6,6 - 8,4 - 4,3 - 5,3 - 3,2 - 7,6 - 4,2 - 8,3 - 2,3 - 6,4

Determine:

a) Calcule a média e o desvio padrão das notas.

b) Dado o intervalo [media - DP, media + DP] determine quantos alunos ficaram dentro desse intervalo, quantos alunos ficaram acima desse intervalo e quantos alunos ficaram abaixo desse intervalo. Use o comando print para expressar as respostas de forma clara.

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as sta
Resultados = [2.3 , 3.1 , 3.1 , 3.2 , 3.7 , 4.2 , 4.3 , 4.4 , 4.8 , 5.0 , 5.3 , 5.4 , 5.6 , 5.9 , 5.9 , 6.4 , 6.6 , 7.6 , 7.8 , 8.3 , 8.4 , 8.5]
media = np.mean(Resultados)
DP = np.std(Resultados)
print(f"Media = {media}")
print(f"DP = {DP}")

Alunos_dentro_do_intervalo_Inferior = media - DP
Alunos_dentro_do_intervalo_superior = media + DP

print(f'Intervalo Inferior: {media - DP}')
print(f'Intervalo Superior: {media + DP}')

# Contando quantos alunos ficaram dentro do intervalo
alunos_dentro_intervalo = sum(1 for nota in Resultados if Alunos_dentro_do_intervalo_Inferior <= nota <= Alunos_dentro_do_intervalo_superior)

# Contando quantos alunos ficaram acima do intervalo
alunos_acima_intervalo = sum(1 for nota in Resultados if nota > Alunos_dentro_do_intervalo_superior)

# Contando quantos alunos ficaram abaixo do intervalo
alunos_abaixo_intervalo = sum(1 for nota in Resultados if nota < Alunos_dentro_do_intervalo_Inferior)

print("Alunos dentro do intervalo:", alunos_dentro_intervalo)
print("Alunos acima do intervalo:", alunos_acima_intervalo)
print("Alunos abaixo do intervalo:", alunos_abaixo_intervalo)
```
